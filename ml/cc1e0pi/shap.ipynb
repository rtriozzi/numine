```python
import numpy
import pandas
import xgboost
import shap

import src.evaluation
```

```python
DATA_PATH = "../data"
MODEL_PATH = "../models"
MODEL_NAME = "cc1e0pi_tagger_xgb"
```

```python
# read the trainied model
model = xgboost.XGBClassifier()
model.load_model(f"{MODEL_PATH}/{MODEL_NAME}.json")

# read pickled training datasets
X_train = pandas.read_pickle(f'{DATA_PATH}/training_dataset_X.pkl')
Y_train = pandas.read_pickle(f'{DATA_PATH}/training_dataset_Y.pkl')
```

```python
shap_values = src.evaluation.compute_shap_values(model, X_train)
```
